```python
import datetime
import numpy as np
import pandas as pd
def lev(str_a,str_b):
    str_a=str_a.lower() # 将引号去掉
    str_b=str_b.lower()
    matrix_ed=np.zeros((len(str_a)+1,len(str_b)+1),dtype=np.int)# 设置一个0矩阵
    matrix_ed[0]=np.arange(len(str_b)+1) # 将第一行赋值
    matrix_ed[:,0] = np.arange(len(str_a) + 1) # 将第一列赋值
    for i in range(1,len(str_a)+1):
        for j in range(1,len(str_b)+1):
            # 表示删除a_i
            dist_1 = matrix_ed[i - 1, j] + 1
            # 表示插入b_i
            dist_2 = matrix_ed[i, j - 1] + 1
            # 表示替换b_i
            dist_3 = matrix_ed[i - 1, j - 1] + (1 if str_a[i - 1] != str_b[j - 1] else 0)
            #取最小距离
            matrix_ed[i,j]=np.min([dist_1, dist_2, dist_3])
    #print(matrix_ed)
    return matrix_ed[-1,-1]
```

```python
start = datetime.datetime.now()
df_1 = pd.read_excel(r'./实验1数据.xlsx',sheet_name='sheet1')
df_2 = pd.read_excel(r'./实验1数据.xlsx',sheet_name='sheet2')
data_1 = df_1.iloc[0:]
data_2 = df_2.iloc[0:]
str_1 = []
str_2 = []
tuple_1 = ()
k = 0
for i in data_1['NAME']:
    k = k + 1
    h = 0
    for j in data_2['name']:
        h = h + 1
        num = lev(i,j)
        if num==0:
            impact_scort = data_1.loc[k-1,"impact_scort"]
            jian_name_1 = data_2.loc[h-1,"jian_name_1"]
            jian_name_2 = data_2.loc[h-1,"jian_name_2"]
            jian_name_3 = data_2.loc[h-1,"jian_name_3"]
            #print(f"sheet1:{i},sheet2:{j},distance={num},impact_scort={impact_scort},jian_name_1={jian_name_1},jian_name_2={jian_name_2},jian_name_3={jian_name_3}")
            tuple_1=(j,jian_name_1,jian_name_2,jian_name_3,impact_scort)
            print(tuple_1)
            str_1.append(tuple_1)
            #print(str_1)
            data_2.drop([h-1])
end = datetime.datetime.now()
print('Running time: %s Seconds'%(end-start))
```
